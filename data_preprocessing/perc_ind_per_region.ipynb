```python
import pandas as pd
```

```python
comp_nac_reg = pd.read_csv(r"C:\Users\pvs31\Desktop\Data Science Project\OP-heist-project\data\combined\company_nace_region.csv",sep=",",encoding="latin-1")

# Create a pivot table to count the companies by 'postal_region' and 'Section'
pivot_table = pd.pivot_table(comp_nac_reg, index='postal_region', columns='Section', values='company_name', aggfunc='count', fill_value=0)

# Reset the index to make 'postal_region' a column
pivot_table.reset_index(inplace=True)

# Calculate the total number of companies in each region
total_companies_per_region = pivot_table.iloc[:, 1:].sum(axis=1)

# Divide each column by the total number of companies in its region
for column in pivot_table.columns[1:]:
    pivot_table[column] = pivot_table[column] / total_companies_per_region

# Rename the columns for clarity (optional)
pivot_table.columns.name = None

print("Values divided by the total number of companies in each region:")
print(pivot_table)

pivot_table.to_csv("../data/combined/percent_ind_per_region.csv")
```
