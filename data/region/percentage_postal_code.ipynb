```python
# Libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/region/workforce_2015.csv", encoding='latin1')
header = df.columns.values.tolist()
# Convert all columns except the first one to integers
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
# Postal code reduction:
def process_col(entry):
    if entry[:5].isdigit():
        return str(entry[:5])
    else:
        return entry
df['Postal code area'] = df['Postal code area'].apply(process_col)
# Optionally, fill NaN values with a specific value (e.g., 0)
df = df.fillna(0)
cleaned_df = df.dropna(how='all')
data = np.array(cleaned_df)
```

```python
#Calculate percentage for each postal code
def cal_percentage(dataset):
    results = []
    for row in dataset:
        row_results = [row[0]]
        current_sum = row[1]
        for value in row[1:]:
            if current_sum == 0:
                row_results.append(1.0)
            else:
                percentage = value / current_sum
                row_results.append(percentage)
        results.append(row_results)
    return results
percentages = cal_percentage(data)
print(percentages)
final = np.vstack((header, percentages))
# Convert the NumPy array to a DataFrame
df = pd.DataFrame(final[1:], columns=final[0])
file_path = "data/region/workforce_2015_percentage.csv"
df.to_csv(file_path, sep=',', index=False)
```
